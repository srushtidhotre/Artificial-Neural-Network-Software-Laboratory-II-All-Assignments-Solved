A3. Write a Python Program using Perceptron Neural Network to recognise even and odd numbers. Given numbers are in ASCII form 0 to 9.

```python
import numpy as np

# Function to convert a digit (0-9) into its 7-bit ASCII binary representation
def ascii_binary_representation(digit):
    return np.array([int(bit) for bit in format(ord(str(digit)), '07b')])

# Step activation function i.e. Sigmoid
def step_function(x):
    return 1 if x >= 0 else 0

# Prepare training data
X = np.array([ascii_binary_representation(i) for i in range(10)])  # Inputs
Y = np.array([1 if i % 2 == 0 else 0 for i in range(10)])          # 1 = even, 0 = odd

# Initialize weights and bias randomly
weights = np.random.rand(7)
bias = np.random.rand(1)
learning_rate = 0.1
epochs = 100

# Perceptron prediction function
def perceptron_predict(x):
    linear_output = np.dot(weights, x) + bias
    return step_function(linear_output)

# Training the perceptron
for epoch in range(epochs):
    for i in range(len(X)):
        y_pred = perceptron_predict(X[i])
        error = Y[i] - y_pred
        weights += learning_rate * error * X[i]
        bias += learning_rate * error

# Testing function
def test_perceptron():
    print("Testing Perceptron for ASCII representation of Digits 0-9:")
    for i in range(10):
        x = ascii_binary_representation(i)
        prediction = perceptron_predict(x)
        result = 'Even' if prediction == 1 else 'Odd'
        print(f"Digit: {i} -> {result}")

# Run test
test_perceptron()
```
