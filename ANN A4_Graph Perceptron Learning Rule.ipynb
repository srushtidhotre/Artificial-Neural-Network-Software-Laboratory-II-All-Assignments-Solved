A4. With a suitable example demonstrate the perceptron learning law with its decision regions using python. Give the output in graphical form.

```python
import numpy as np
import matplotlib.pyplot as plt

# Input data: 2 classes
X = np.array([[1,2], [2,3], [3,3], [5,2], [6,3], [7,2]])
y = np.array([0, 0, 0, 1, 1, 1])  # labels 0 and 1

# Convert 0 -> -1 for perceptron math
y = np.where(y == 0, -1, 1)

# Add bias column (1s) to input
Xb = np.hstack((np.ones((X.shape[0], 1)), X))

# Initialize weights to small random values instead of zeros
w = np.random.uniform(-0.5, 0.5, Xb.shape[1])
learning_rate = 0.1

# Train perceptron with print per epoch
for epoch in range(10):
    errors = 0
    for xi, target in zip(Xb, y):
        output = np.sign(np.dot(xi, w))
        if output != target:
            w += learning_rate * (target - output) * xi
            errors += 1
    print(f"Epoch {epoch+1}: Errors = {errors}")

print("Final weights:", w)

# Plot points and decision boundary
plt.scatter(X[y==1][:,0], X[y==1][:,1], color='blue', label='Class 1')
plt.scatter(X[y==-1][:,0], X[y==-1][:,1], color='red', label='Class -1')
x_line = np.linspace(0,8,100)
y_line = -(w[0] + w[1]*x_line) / w[2]
plt.plot(x_line, y_line, linestyle='--', label='Decision Boundary')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.title('Perceptron Decision Boundary (with Random Init)')
plt.show()
```
