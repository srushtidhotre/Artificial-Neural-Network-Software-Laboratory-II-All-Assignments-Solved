A2. Generate ANDNOT function using McCulloch-Pitts neural net by a python program.

```python
import numpy as np
inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
outputs = np.array([0, 0, 1, 0])

weights = np.array([1, -1])
threshold = 1

predictions = np.zeros(len(inputs))

for i in range(len(inputs)):
    weighted_sum = np.dot(weights, inputs[i])
    if weighted_sum >= threshold:
        predictions[i] = 1  # Neuron fires
    else:
        predictions[i] = 0  # Neuron does not fire

print("Input (A, B)   Predicted Output   Actual Output")
for i in range(len(inputs)):
    print(f"   {inputs[i][0]}    {inputs[i][1]}            {int(predictions[i])}               {outputs[i]}")
```

```python
def mcCullochPitts_ANDNOT(x1, x2):
    w1 = 1
    w2 = -1
    theta = 1
    net_input = x1 * w1 + x2 * w2
    if net_input >= theta:
        return 1
    else:
        return 0

inputs = [(0,0), (0,1), (1,0), (1,1)]
print("x1 x2 | Output")
for x1, x2 in inputs:
    output = mcCullochPitts_ANDNOT(x1, x2)
    print(f"{x1}  {x2}  |   {output}")
```
