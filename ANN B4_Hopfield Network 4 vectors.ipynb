B4. Write a python program to design a Hopfield Network which stores 4 vectors.

```python
import numpy as np

# Step 1: Initialize the network parameters
size = 4                                                # Size of each pattern
patterns = [
              [1, -1, 1, -1],                           # Pattern 1
              [-1, 1, 1, -1],                           # Pattern 2
              [1, 1, -1, -1],                           # Pattern 3
              [-1, -1, 1, 1]                            # Pattern 4
]

# Step 2: Initialize the weights matrix
weights = np.zeros((size, size))

# Step 3: Train the network (Hebbian learning rule)
for pattern in patterns:
    pattern = np.array(pattern).reshape(size, 1)                                # Reshape to column vector
    weights += np.dot(pattern, pattern.T)                                       # Hebbian update rule

np.fill_diagonal(weights, 0)                                                    # Remove the diagonal (no self-feedback)

# Step 4: Define the predict function
def predict(input_pattern, max_iterations=10):

    input_pattern = np.array(input_pattern).reshape(size, 1)

    for _ in range(max_iterations):

        # Calculate the weighted sum of the inputs
        output = np.sign(np.dot(weights, input_pattern))

        # If there's no change in the output, break the loop
        if np.array_equal(output, input_pattern):
            break

        input_pattern = output

    return output.flatten()

# Step 5: Test the network with a noisy version of a stored pattern
input_pattern = [1, -1, 1, 1]  # A noisy version of pattern1
output_pattern = predict(input_pattern)

# Step 6: Display the results
print("Input Pattern: ", input_pattern)
print("Output Pattern (Retrieved): ", output_pattern)
```
