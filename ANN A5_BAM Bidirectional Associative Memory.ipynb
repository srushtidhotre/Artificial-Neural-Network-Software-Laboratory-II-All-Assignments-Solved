A5. Write a python program for bidirectional associative memory with two pairs of vectors

```python
wimport numpy as np

class BidirectionalAssociativeMemory:
    def __init__(self, input_size, output_size):
        self.input_size = input_size
        self.output_size = output_size
        self.weights = np.zeros((input_size, output_size))

    def train(self, input_vector, output_vector):
        # Update the weight matrix using the outer product
        self.weights += np.outer(input_vector, output_vector)

    def recall_input(self, output_vector):
        # Recalling input vector from output vector using weights
        input_vector = np.sign(np.dot(output_vector, self.weights))
        return input_vector

    def recall_output(self, input_vector):
        # Recalling output vector from input vector using weights
        output_vector = np.sign(np.dot(input_vector, self.weights.T))
        return output_vector

# Define input-output vector pairs
input1 = np.array([1, -1, 1])
input2 = np.array([-1, 1, 1])
output1 = np.array([1, 1, -1])
output2 = np.array([-1, -1, 1])

# Initialize BAM
bam = BidirectionalAssociativeMemory(input_size=3, output_size=3)

# Train BAM with two input-output pairs
bam.train(input1, output1)
bam.train(input2, output2)

# Test BAM by recalling
recalled_output1 = bam.recall_output(input1)
recalled_input1 = bam.recall_input(output1)

recalled_output2 = bam.recall_output(input2)
recalled_input2 = bam.recall_input(output2)

print("Recalled Output for Input1:", recalled_output1)
print("Recalled Input for Output1:", recalled_input1)
print("Recalled Output for Input2:", recalled_output2)
print("Recalled Input for Output2:", recalled_input2)
```
