A1. Write a Python program to plot a few activation functions that are being used in neural networks.

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid Activation Function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# ReLU Activation Function
def relu(x):
    return np.maximum(0, x)

# Tanh Activation Function
def tanh(x):
    return np.tanh(x)

# Softmax Activation Function (typically used in the output layer for classification)
def softmax(x):
    e_x = np.exp(x - np.max(x))  # Subtracting np.max(x) for numerical stability
    return e_x / e_x.sum(axis=0, keepdims=True)

# Plotting the activation functions
def plot_activation_functions():
    x = np.linspace(-10, 10, 100)

    # Generate outputs for each activation function
    sigmoid_output = sigmoid(x)
    relu_output = relu(x)
    tanh_output = tanh(x)

    # Plot each activation function
    plt.figure(figsize=(9, 6))

    # Sigmoid Plot
    plt.subplot(2, 2, 1)
    plt.plot(x, sigmoid_output, label="Sigmoid")
    plt.title("Sigmoid Activation Function")
    plt.legend()

    # ReLU Plot
    plt.subplot(2, 2, 2)
    plt.plot(x, relu_output, label="ReLU")
    plt.title("ReLU Activation Function")
    plt.legend()

    # Tanh Plot
    plt.subplot(2, 2, 3)
    plt.plot(x, tanh_output, label="Tanh")
    plt.title("Tanh Activation Function")
    plt.legend()

    # Softmax Plot
    softmax_input = np.array([[x_val, 0, 0] for x_val in x])
    softmax_output = softmax(softmax_input)
    plt.subplot(2, 2, 4)
    plt.plot(x, softmax_output[:,0], label="Softmax Output (class 1)")
    plt.title("Softmax Activation Function")
    plt.legend()

    # Show all plots
    plt.tight_layout()
    plt.show()

# Run the function to plot activation functions
if __name__ == "__main__":
    plot_activation_functions()
```
